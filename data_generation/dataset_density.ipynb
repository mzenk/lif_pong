```python
from __future__ import division, print_function
import numpy as np
from scipy.stats import gaussian_kde
import matplotlib.pyplot  as plt
%matplotlib inline
```

```python
def plot_histo(pos, angles, bins=10, title=''):
    init_hist, xedges, yedges = np.histogram2d(pos, angles, bins=bins)
    # plotting
    fig_hist, ax_hist = plt.subplots(figsize=(10, 7))
    ax_hist.set(xlabel='position', ylabel='angle')
    ax_hist.set_title(title)
    X, Y = np.meshgrid(xedges, yedges)
    hist_im = ax_hist.pcolormesh(X, Y, init_hist.T)
    plt.colorbar(hist_im, ax=ax_hist)
    
def plot_kde(pos, angles):
    xmin = pos.min()
    xmax = pos.max()
#     ymin = angles.min()
#     ymax = angles.max()
    ymin = -90.
    ymax = 90.
    X, Y = np.meshgrid(np.linspace(xmin, xmax, 50), np.linspace(ymin, ymax, 50))
    
    kernel = gaussian_kde(np.vstack((pos, angles)))
    positions = np.vstack((X.ravel(), Y.ravel()))
    Z = np.reshape(kernel(positions).T, X.shape)
    
    fig_kde, ax_kde = plt.subplots(figsize=(10, 7))
    ax_kde.set(xlabel='position', ylabel='angle')
    im = ax_kde.pcolormesh(X, Y, Z)
    plt.colorbar(im, ax=ax_kde)
```

```python
data_stats = np.load('pong_lw5_stats.npz')
init_pos = data_stats['init_pos']
init_angles = data_stats['init_angles']
end_pos = data_stats['end_pos']
print(init_pos.min(), init_pos.max(), init_angles.min(), init_angles.max())

ntrain = 20000
ntest = 10000

train_pos = init_pos[:ntrain]
train_angles = init_angles[:ntrain]
test_pos = init_pos[-ntest:]
test_angles = init_angles[-ntest:]


# plot_histo(train_pos, train_angles, [20, 20])
# plot_histo(test_pos, test_angles, [20, 20])

for n in range(2000, ntrain + 1, 2000):
    plot_histo(init_pos[:n], init_angles[:n], [20, 20], title='N={}'.format(n))
```

```python
plot_kde(train_pos, train_angles)
plot_kde(test_pos, test_angles)
```
